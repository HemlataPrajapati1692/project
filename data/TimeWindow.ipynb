```python
import pandas as pd
from pandas.io import sql
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sqlite3
import pandas_profiling

%matplotlib inline
```

```python
conn = sqlite3.connect('EuropeanSoccerDatabase/database.sqlite')
query = "SELECT team.id, team.team_api_id, team.team_long_name, match.league_id FROM Team INNER JOIN Match on team.team_api_id = match.home_team_api_id OR team.team_api_id = match.away_team_api_id WHERE Match.league_id = 7809 Group by team_api_id  ;"

Teams = sql.read_sql(query, con=conn, parse_dates="date")
```

```python
id_homeTeam = 9789
id_awayTeam = 9823
Matchdate = np.datetime64('2009-11-04')
```

```python
conn = sqlite3.connect('EuropeanSoccerDatabase/database.sqlite')
query = "SELECT * FROM match WHERE ( home_team_api_id ="+str(id_homeTeam)+" OR away_team_api_id="+str(id_homeTeam)+") AND date <'"+ str(Matchdate)+"' ORDER BY date DESC LIMIT 10 ;"

bundesliga_matchs_homeTeam = sql.read_sql(query, con=conn, parse_dates="date")
```

```python
conn = sqlite3.connect('EuropeanSoccerDatabase/database.sqlite')
query = "SELECT * FROM match WHERE ( home_team_api_id ="+str(id_awayTeam)+" OR away_team_api_id="+str(id_awayTeam)+") AND date <'"+ str(Matchdate)+"' ORDER BY date DESC LIMIT 10;"

bundesliga_matchs_awayTeam = sql.read_sql(query, con=conn, parse_dates="date")
```
