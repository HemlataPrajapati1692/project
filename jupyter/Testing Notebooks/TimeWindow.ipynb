```python
def pointsForHomeTeam(row):
    if row['home_team_goal'] > row['away_team_goal']:
        return 3 #win : 3pts
    if row['home_team_goal'] < row['away_team_goal']:
        return 0 #lose : 0pt
    else:
        return 1 #draw : 1pt

def pointsForAwayTeam(row):
    if row['home_team_goal'] < row['away_team_goal']:
        return 3 #win : 3pts
    if row['home_team_goal'] > row['away_team_goal']:
        return 0 #lose : 0pt
    else:
        return 1 #draw : 1pt
```

```python
bundisliga_matchs
```

```python
bundisliga_matchs['home_team_earned_points'] = bundisliga_matchs.apply(lambda row: pointsForHomeTeam(row), axis=1)
bundisliga_matchs['away_team_earned_points'] = bundisliga_matchs.apply(lambda row: pointsForAwayTeam(row), axis=1)
bundisliga_matchs
```

```python
filt1 = (bundisliga_matchs['away_team_api_id'] == id_homeTeam)
away_team_points = bundisliga_matchs[filt1].groupby('away_team_api_id').sum()[['away_team_earned_points']]

blub1 = away_team_points.iloc[0,0]

filt2 = (bundisliga_matchs['home_team_api_id'] == id_homeTeam)
home_team_points = bundisliga_matchs[filt2].groupby('home_team_api_id').sum()[['home_team_earned_points']]

blub2 = home_team_points.iloc[0,0]

pointsHomeTeam = blub1 + blub2

filt3 = (bundisliga_matchs['away_team_api_id'] == id_awayTeam)
away_team_points = bundisliga_matchs[filt3].groupby('away_team_api_id').sum()[['away_team_earned_points']]

blub1 = away_team_points.iloc[0,0]

filt4 = (bundisliga_matchs['home_team_api_id'] == id_awayTeam)
home_team_points = bundisliga_matchs[filt4].groupby('home_team_api_id').sum()[['home_team_earned_points']]

blub2 = home_team_points.iloc[0,0]

pointsAwayTeam = blub1 + blub2
```

```python
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1))
```
